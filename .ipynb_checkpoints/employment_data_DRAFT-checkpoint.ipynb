```python
import os
import datetime
import requests
import json
import pandas as pd
import hvplot.pandas
import plotly.express as px
%matplotlib inline
```

```python
headers = {'Content-type': 'application/json'}
data = json.dumps({"seriesid": ['CEU3231160001'],"startyear":"2019", "endyear":"2020", "registrationkey":"3040f2eebdb244b5b9ce84ba21f0c6c9"})
p = requests.post('https://api.bls.gov/publicAPI/v2/timeseries/data/', data=data, headers=headers)
json_data = json.loads(p.text)
```

```python
list_of_dictionaries = json_data['Results']['series'][0]['data'] #list of dictionaries
list_of_dictionaries #everything is a string... to do numerical analysis you need NUMBERS
```

```python
year_list=[]
month=[]
value=[]

for items in json_data['Results']['series'][0]['data']:
    year_list.append(items['year']) #should change to an integer but not necessary yet
    month.append(items['periodName']) #should be a # between 1-12, but not neccesary with the datatime regex functions
    value.append(float(items['value'])) #has to be a #
```

```python
year_df = pd.DataFrame([year_list]).transpose()
month_df = pd.DataFrame([month]).transpose()
value_df = pd.DataFrame([value]).transpose()
```

```python
employment_data_df = pd.concat([year_df, month_df, value_df], axis='columns', join='inner')
employment_data_df
```

```python
employment_data_df.columns = ["Year", "Month", "Value"]
employment_data_df
#place as datetime. it was already done. panda sees it datewith time and seperate
employment_data_df['date'] = pd.to_datetime(employment_data_df['Year'] + employment_data_df['Month'], format='%Y%B') #2020March and they are already strings
```

```python
employment_data_df
```

```python
# no need to sert index to dateframe already done. change to 'date'
employment_data_df.set_index('date', inplace=True)
employment_data_df
```

```python
employment_data = employment_data_df.drop(columns = ['Year','Month'])
employment_data
```

```python
employment_data.plot.line(title = 'Employees in Animal Slaughtering and Processing Industry').set_ylabel('Employees, Thousands')
```

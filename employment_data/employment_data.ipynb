```python
import os
import datetime
import requests
import json
import pandas as pd
import hvplot.pandas
import plotly.express as px
import panel as pn
import matplotlib.pyplot as plt
%matplotlib inline
pn.extension('plotly')
```

```python
headers = {'Content-type': 'application/json'}
data = json.dumps({"seriesid": ['CEU3231160001'],"startyear":"2019", "endyear":"2020", "registrationkey":"3040f2eebdb244b5b9ce84ba21f0c6c9"})
p = requests.post('https://api.bls.gov/publicAPI/v2/timeseries/data/', data=data, headers=headers)
json_data = json.loads(p.text)
```

```python
list_of_dictionaries = json_data['Results']['series'][0]['data']
list_of_dictionaries
```

```python
year_list=[]
month=[]
value=[]

for items in json_data['Results']['series'][0]['data']:
    year_list.append(items['year'])
    month.append(items['periodName'])
    value.append(float(items['value']))
```

```python
year_df = pd.DataFrame([year_list]).transpose()
month_df = pd.DataFrame([month]).transpose()
value_df = pd.DataFrame([value]).transpose()
```

```python
employment_data_df = pd.concat([year_df, month_df, value_df], axis='columns', join='inner')
employment_data_df
```

```python
employment_data_df.columns = ["Year", "Month", "Value"]
employment_data_df
employment_data_df['date'] = pd.to_datetime(employment_data_df['Year'] + employment_data_df['Month'], format='%Y%B')
```

```python
employment_data_df
```

```python
employment_data_df.set_index('date', inplace=True)
employment_data_df
```

```python
employment_data = employment_data_df.drop(columns = ['Year','Month'])
employment_data
```

```python
employment_data.hvplot.line(label = 'Employees in Animal Slaughtering and Processing Industry', ylabel= 'Employees, Thousands')
```

```python
def employment_data_hours_avg():
    "Average Weekly Hours of Employees in Animal Slaughtering & Processing Industry."
    
    #suggest putting all work here for employment_data_hours df
    
    employment_data_hours_panel = employment_data_hours.hvplot.line(label = 'Average Weekly Hours of Employees in Animal Slaughtering & Processing Industry', 
                                                                    ylabel='Hours',
                                                                    ylim = (employment_data_hours['Value'].min()*.99,
                                                                            employment_data_hours['Value'].max()*1.01
                                                                           )
                                                                   
                                                                   
                                                                   ).opts(axiswise=True)
    
    return employment_data_hours_panel
```
